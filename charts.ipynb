```python
import seaborn as s
import matplotlib.pyplot as p
```

```python
tip=s.load_dataset("tips")
tip.head()
```

```python
s.scatterplot(x="day",y="total_bill",data=tip)
p.title("Scatter Plot")
p.xlabel("Tip")
p.ylabel("Size")
p.show()
```

```python
s.barplot(x="day",y="total_bill",data=tip)
p.title("bar Plot")
p.xlabel("Tip")
p.ylabel("Size")
p.show()
```

```python
s.lineplot(x="day",y="total_bill",data=tip)
p.title("LIne Plot")
p.xlabel("Tip")
p.ylabel("Size")
p.show()
```

```python
s.displot(x="day",y="total_bill",data=tip)
p.title("dis Plot")
p.xlabel("Tip")
p.ylabel("Size")
p.show()
```

```python
s.boxplot(x="day",y="total_bill",data=tip)
p.title("BoxPlot")
p.xlabel("Tip")
p.ylabel("Size")
p.show()
```

```python
s.histplot(x="day",y="total_bill",data=tip)
p.title("hist Plot")
p.xlabel("Tip")
p.ylabel("Size")
p.show()
```

```python
s.violinplot(x="day",y="total_bill",data=tip)
p.title("violin Plot")
p.xlabel("Tip")
p.ylabel("Size")
p.show()
```

```python
iris=s.load_dataset("iris")
correlation_matr=iris.drop('species', axis=1).corr()
s.heatmap(correlation_matr,annot=True,cmap="coolwarm")
p.title("Correlation Heatmap")
p.show()
```

```python
s.countplot(x="day",data=tip)
p.title("Count Plot")
p.xlabel("Tip")
p.ylabel("Size")
p.show()
```

```python
s.pairplot(tip,hue="tip")
p.title("Pair Plot")
p.xlabel("Tip")
p.ylabel("Size")
p.show()
```

```python
s.pointplot(x="day",y="total_bill",data=tip)
p.title("Point Plot")
p.xlabel("Tip")
p.ylabel("Size")
p.show()
```
